```python
# put the needed libraries first, here at the top
import numpy as np # math
import matplotlib.pyplot as plt # graphing

#Knowns - what variables did the problem give us? don't forget units!
x  = float(input("How high is the ball dropped from? ")) # get input from user and typecast it, meters

#in addition to the known values, are there any assumed values?

t = 0 # seconds
v = 0 # meters / second
a = -9.8 # meters / second squared, g

#We also want to update time incrementally with the Euler-Cromer method
dt = 0.00001 # time step

#Lists - these will typically be used to create graphs, so think about what we want to graph! Don't forget units

tlist = [t] #s
xlist = [x] #m
vlist = [v] #v

#The Motion - Using a while loop, we update the various variables of the object

while x > 0: #example - while t < 10:
    v += a * dt #what physics is happening here? v = a * t
    x += v * dt #what physics is happening here? x = v * t
    t += dt #what physics is happening here? update time

    #we also need to update our lists
    tlist.append(t) # add time to list
    xlist.append(x) # add height to list
    vlist.append(v) # add velocity to list


#Graphs - what are the common physics graphs you'd be asked for a question like this? distance vs. time, velocity vs. time

fig, axs = plt.subplots(2)
fig.suptitle("Ball Drop Graphs")

axs[0].scatter(tlist, xlist, color='blue');   # this makes a scatterplot!
axs[0].set_xlabel("Time (s)") # x is time
axs[0].set_ylabel("Height (m)") # y is height
axs[0].set_title("Ball Height vs. Time") # title

axs[1].scatter(tlist, vlist, color='purple');   # this makes a scatterplot!
axs[1].set_xlabel("Time (s)") # x is time
axs[1].set_ylabel("Velocity (m/s)") # y is velocity
axs[1].set_title("Ball Velocity vs. Time") # title

plt.tight_layout()
plt.show()

#How do we get the computer to tell us the answer? Where is this variable being stored? How do we print it out?
print(f"The ball takes {t} seconds to hit the ground when dropped from {xlist[0]} meters, with a final velocity of {v} meters per second") 
    # show time taken and input height 
```
