# Part 1: How do we calculate acceleration?

Take the example question: A car has a mass of 1500 kg. If the driver applies the brakes while on a gravel road, the maximum friction force that the tires can provide without skidding is about 7000 N. If the car is moving at 20 m/s, what is the shortest distance in which the car can stop safely? (Knight Ch. 4, problem 62)


```python
# put the needed libraries first, here at the top
import numpy as np
import matplotlib.pyplot as plt

#Knowns - what variables did the problem give us? don't forget units!
m = 1500 # kg
F_friction = -7000 # newtons
v = 20 # meters / second

#in addition to the known values, are there any assumed values?

t = 0 # seconds
x = 0 # meters
a = F_friction / m #what physics is happening here? a = F/m

#We also want to update time incrementally with the Euler-Cromer method
dt = 0.0001 # time step

#Lists - these will typically be used to create graphs, so think about what we want to graph! Don't forget units

tlist = [t] #s
xlist = [x] #m
vlist = [v] #v

#The Motion - Using a while loop, we update the various variables of the object

while v > 0: #example - while t < 10:
    v += a * dt #what physics is happening here? v = a * t
    x += v * dt #what physics is happening here? x = v * t
    t += dt #what physics is happening here? update time

    #we also need to update our lists
    tlist.append(t)
    xlist.append(x)
    vlist.append(v)


#Graphs - what are the common physics graphs you'd be asked for a question like this? distance vs. time, velocity vs. time
#Do you have a graphing code you can copy and paste here?
plt.scatter(tlist, xlist, color='blue');   # this makes a scatterplot!
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.title("Car Distance vs. Time")

plt.show() #this shows the graph

plt.scatter(tlist, vlist, color='purple');   # this makes a scatterplot!
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.title("Car Speed vs. Time")

plt.show() #this shows the graph

#How do we get the computer to tell us the answer? Where is this variable being stored? How do we print it out?
print(f"The minimum safe distance that the car can stop at is {x} meters")
```

Now your turn. Try this problem out: A stationary car with a mass of 1500 kg is being towed by a rope pulling with 2300 N held at a 20 degree angle to the horizontal. How long does it take the car to reach 90 m? (modified from example 5.6, Knight)

```python
#We don't have to put the libraries in this code if the coding cell above has already been run this session
import numpy as np
import matplotlib.pyplot as plt

#Knowns
m = 1500 # kg
F_pull = 2300 # newtons
theta = 20 # degrees

t = 0 # seconds
v = 0 # meters / seconds
x = 0 # meters

a = F_pull * np.cos(20 * np.pi / 180) / m #what physics is happening here? a = F/m

dt = 0.0001

#Lists
tlist = [t] #s
xlist = [x] #m
vlist = [v] #v

#The Motion
while x < 90: 
    
    v += a * dt #what physics is happening here? v = a * t
    x += v * dt #what physics is happening here? x = v * t
    t += dt #what physics is happening here? update time

    #we also need to update our lists
    tlist.append(t)
    xlist.append(x)
    vlist.append(v)

#Graphs
plt.scatter(tlist, xlist, color='indigo');   # this makes a scatterplot!
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.title("Car Distance vs. Time")

plt.show() #this shows the graph

plt.scatter(tlist, vlist, color='maroon');   # this makes a scatterplot!
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.title("Car Speed vs. Time")

plt.show() #this shows the graph

#How do we get the computer to tell us the answer? Print it out!
print(f"The car takes {t} seconds to reach {x} meters")
```
