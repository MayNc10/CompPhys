```python
def model_function(x): # Defines a function that takes an x value and returns a y value
    a = 2
    b = -7
    y = a * (x**2) + b
    return y

xvals = np.linspace(0, 5, num=10) # Generates 5 numbers between 0 and 10, including both endpoints
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
yvals = model_function(xvals) # Computes the y values for each x value
plt.scatter(xvals, yvals) # Creates a scatter plot with the x and y values
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
```

Below is another line of code that uses the same data but plots differently. Run the code and determine what the difference is.

#### Next step: Customization

The preivous two graphs use the pyplot interface. This interface can generate quick and easy plots, but is best for simple plots.

The code below is an example of using the object oriented interface. It creates a figure (fig) that can hold many different elements, allowing for greater customization.

For each line of code, describe what the line does. If there are multiple features in a line of code, describe each element. Again, feel free to manipulate variables to aid in your exploration.

```python
fig, ax = plt.subplots(2, 2, figsize=(10,5)) # Create a general figure, it can be reused for making different subplots
fig.subplots_adjust(hspace=0.5) # Adjust height padding between subplots

ax[0,0].plot(xvals, yvals, color= 'chartreuse', label='one', linestyle="--") # Create a dashed line plot of x and y values
ax[0,0].plot(xvals, xvals*2+4, color= 'skyblue', label='two', linestyle="--") 
# Create a dashed line plot of the x and y values using the x value data, with some scaling and shifting

ax[0,0].set(xlabel="numbers", ylabel="more numbers", title="the numbers") # Set axis labels and graph title
idx = np.argwhere(np.diff(np.sign(yvals - (xvals*2+4)))).flatten() # find where the two graphs intersect
ax[0,0].annotate('there is an intersection here',xy=(xvals[idx], yvals[idx]), xytext=(xvals[idx] - 2, yvals[idx] + 10), 
    arrowprops=dict(facecolor='black', arrowstyle="->")) # Add an arrow pointing to the intersection point

ax[0,0].legend(loc="upper center") # Adds a legend in the upper center

ax[1,0].set_title("this is a title") # Sets a title
ax[1,0].annotate('thats my name!',xy=(.5, .5), xytext=(.2, .7), 
    arrowprops=dict(facecolor='black', arrowstyle='fancy')) # Create an annotation and an arrow
ax[1,0].plot([.5], [.5], marker='$may$', mfc='#4287f5', mec='#4287f5') # Plot a marker
ax[1,0].set_xlim(-5,5) # Set x limit
ax[1,0].set_ylim(-5,5) # Set y limit

ax[1,0].spines['top'].set_visible(False) # Remove the top box line
ax[1,0].spines['right'].set_visible(False) # Remove the right box line
ax[1,0].spines['bottom'].set_color('#565656') # Set the color of the bottom box line

ax[0,1].set(xlabel="some label", ylabel="other label", title="a title?") # Set axis labels and graph title
ax[0,1].set_xlim(2,5) # Adjusts the shown x values
ax[0,1].xaxis.set_ticks_position('none') # Remove tick lines on the x axis
ax[0,1].tick_params(labelcolor='grey') # Make the tick labels grey
ax[0,1].grid(color='#565656', axis='y', linestyle='-', linewidth=0.25, alpha=0.5) # Change the color and alpha of the grid lines on the y axis
ax[0,1].grid(color='#B00B69', axis='x', linestyle='-', linewidth=0.25, alpha=1) # Change the color and alpha of the grid lines on the x axis
ax[0,1].set_xscale("log") # Make the x axis a log scale
ax[0,1].scatter(xvals, yvals, color=['#7ffc03'], label='log', marker='$>:)$') # scatter plot the data

import random # Import the random library

science_things = np.linspace(random.randint(-1000, -1), random.randint(1, 1000), random.randint(10, 100), bool(random.getrandbits(1))) 
    # Create a random range of numbers
def random_func(x):
    a = random.randint(-5, 5)
    b = random.randint(-100, 100)
    c = random.randint(-100, 100)
    return c + b * (x ** a) # Create an exponential function with random exponent, scaling, and shift up/down
science_results = random_func(science_things) # get our results
ax[1,1].plot(science_things, science_results, 'pink', alpha=1, label='stuff', marker="$:)$", mfc='#03dbfc', mec='#03dbfc') # plot data
ax[1,1].set(xlabel="things", ylabel="results", title="science!") # add labels and title
ax[1,1].grid(color='orange', axis='y', linestyle='-', linewidth=0.25) # add y grid line
ax[1,1].grid(color='purple', axis='x', linestyle='-', linewidth=0.25) # add x grid line
ax[1,1].text(science_things[len(science_things)//2], science_results[len(science_results)//2], "i dont know what this looks like!") # Add text on the graph


plt.show() # Display the graphs
```
