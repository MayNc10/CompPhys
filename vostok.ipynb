```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("vostok data.csv")
data.head(3)

plt.scatter(-data["Age of ice (yr BP)"], -data["Depth (m)"])
plt.xlabel("Age of ice")
plt.ylabel("Depth")
plt.show()

plt.scatter(-data["Mean age of air (yr BP)"], data["CO2 concentration (ppmv)"])
plt.xlabel("Age of air")
plt.ylabel("CO2 concentration")
plt.show()

```
