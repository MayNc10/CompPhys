```python
# put the needed libraries first, here at the top
import numpy as np # math
import matplotlib.pyplot as plt # graphing

#knows
m = 1 # kg
g = np.array([0, -9.8]) # m / s^2, gravity

pos = np.array([0,0]) # m
v = np.array([35, 27]) # m / s

t = 0 # s
dt = 0.1 # time step

ylist = [pos[1]]
xlist = [pos[0]]
vylist = [v[1]]
vxlist = [v[0]]
tlist = [t]

#while loop
while ylist[-1] >= 0:
    Fnet = m * g
    a = Fnet / m

    v = v + a * dt
    pos = pos + v * dt
    t += dt

    ylist.append(pos[1])
    xlist.append(pos[0])

    vylist.append(v[1])
    vxlist.append(v[0])

    tlist.append(t)

fig, axs = plt.subplots(3)
fig.suptitle("Ball Drop Graphs")

axs[0].scatter(tlist, ylist, color='blue');   # this makes a scatterplot!
axs[0].set_xlabel("Time (s)") # x is time
axs[0].set_ylabel("Height (m)") # y is height
axs[0].set_title("Ball Height vs. Time") # title

axs[1].scatter(tlist, xlist, color='blue');   # this makes a scatterplot!
axs[1].set_xlabel("Time (s)") # x is time
axs[1].set_ylabel("Distance (m)") # y is height
axs[1].set_title("Ball Distance vs. Time") # title

axs[2].scatter(xlist, ylist, color='purple');   # this makes a scatterplot!
axs[2].set_xlabel("Distance (s)") # x is time
axs[2].set_ylabel("Height (m/s)") # y is velocity
axs[2].set_title("Ball Height vs. Distance") # title


plt.tight_layout()
plt.show()

```
