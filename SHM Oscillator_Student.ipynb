```python
#define the object parameters
m = 1 #kg mass of block on horizontal surface
k = 25 #N/m spring constant
L = 1 #m unstretched length of spring

#define simulation parameters
x0 = 0.5 #m initial position of the block
v0 = 0 #m/s initial velocity of the block
t0 = 0 #s initial time
dt = 0.01 #s simulation time step
sim_length = 10 #s simulation length

#initialize states
x=x0
v=v0
t=t0

#initialize lists for graphing later
x_list=[x]
v_list=[v]
t_list=[t]

while t<sim_length:
    #calculate stuff
    F_spring = -k * (x - L)
    F_net = F_spring
    a = F_net / m
    v += a * dt
    x += v * dt
    t += dt
    #append lists
    x_list.append(x)
    v_list.append(v)
    t_list.append(t)
    
import matplotlib.pyplot as plt
plt.plot(t_list,x_list)
plt.show()
```

```python
import numpy as np

#Let's add some more forces
#define the object parameters
m = 1 #kg mass of block on horizontal surface
g = -9.8
k = 25 #N/m spring constant
L = 1 #m unstretched length of spring

mu = 0.2

#define simulation parameters
x0 = 0.5 #m initial position of the block
v0 = 0 #m/s initial velocity of the block
t0 = 0 #s initial time
dt = 0.01 #s simulation time step
sim_length = 10 #s simulation length

#initialize states
x=x0
v=v0
t=t0

#initialize lists for graphing later
x_list=[x]
v_list=[v]
t_list=[t]

while t<sim_length:
    #calculate stuff
    F_spring = -k * (x - L)
    F_fric = -np.sign(v) * mu * m * -g
    F_net = F_spring + F_fric
    a = F_net / m
    v += a * dt
    x += v * dt
    t += dt
    #append lists
    x_list.append(x)
    v_list.append(v)
    t_list.append(t)
    
import matplotlib.pyplot as plt
plt.plot(t_list,x_list)
plt.show()
```

```python
#What if the spring was a vertical oscillator?
#define the object parameters
m = 1 #kg mass of block on horizontal surface
k = 25 #N/m spring constant
L = 1 #m unstretched length of spring
g = 9.8

#define simulation parameters
y0 = 2 #m initial position of the block
v0 = 0 #m/s initial velocity of the block
t0 = 0 #s initial time
dt = 0.01 #s simulation time step
sim_length = 10 #s simulation length

#initialize states
y=y0
v=v0
t=t0

#initialize lists for graphing later
y_list=[y]
v_list=[v]
t_list=[t]

while t<sim_length:
    #calculate stuff
    F_spring = -k * (y - L)
    F_g = m * g
    F_net = F_spring - F_g
    a = F_net / m
    v += a * dt
    y += v * dt
    t += dt
    #append lists
    y_list.append(y)
    v_list.append(v)
    t_list.append(t)
    
import matplotlib.pyplot as plt
plt.plot(t_list,y_list)
plt.show()
```
