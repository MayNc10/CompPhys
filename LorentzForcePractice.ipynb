```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters for plot attributes
plt.rc("xtick", labelsize="large")
plt.rc("ytick", labelsize="large")
plt.rc("axes", labelsize="xx-large")
plt.rc("axes", titlesize="xx-large")
plt.rc("figure", figsize=(8,8))


# define key constants
m_p = 1.67E-27     # mass of proton: kg
m_e = 9.1E-31
qe =  1.6022E-19      # charge of proton: C

# now, setting up the particle
m = m_e  #kg
q = -qe #C

dt = 3.0E-8 #s
t = 0 #s starting time

pos = np.array([0,0,0]) #m, sets initial position
v = np.array([6E+7, 0, 0]) #m/s, sets initial velocity

#create lists for graphing later
xlist = [pos[0]]
ylist = [pos[1]]
zlist = [pos[2]]
tlist = [t]


# Strength of Fields
B = np.array([0, 0, 5E-5]) #Unit: Teslas, magnetic field is in the z-direction 
#E = np.array([0, 0, 0]) #unit: N/C, electric field is zero for this example
E = -np.cross(v, B)

# Euler time steps
while t < 0.01:
    Fnet = q * (E + np.cross(v, B)) #Defining net force vector
    a = Fnet / m #calculating accerlation vector, Newton's 2nd Law
    v = v + a * dt #updating velocity vector 
    pos = pos + v * dt #updating position vector
    t += dt #updating time
    
    #updating lists
    xlist.append(pos[0])
    ylist.append(pos[1])
    zlist.append(pos[2])
    tlist.append(t)
    
# Plot the particle's trajectory, in the xy-plane
plt.plot(xlist,ylist, label="Alpha particle")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Particle Trajectories in Uniform Magnetic Field")
plt.legend(fontsize="x-large")
plt.xlim(-13,13)
plt.ylim(-13,13)
plt.show()

#Now plot the particle's trajector in the xz-plane. What do you notice?
plt.plot(xlist,zlist, label="Alpha particle")
plt.xlabel("$x$")
plt.ylabel("$z$")
plt.title("Particle Trajectories in Uniform Magnetic Field")
plt.show()

print(f"Electric field strength: {E} N/C")
```
