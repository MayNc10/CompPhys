```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# creating a coordinate system for the field
x = np.linspace(0, 10, 10)
y = np.linspace(0, 10, 10)

X,Y = np.meshgrid(x,y)

u = 1/(np.cos(X))
v = 1/(np.sin(Y))

plt.quiver(X, Y, u, v)
plt.title("Gravity")
plt.show()
```
