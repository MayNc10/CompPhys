```python
# put the needed libraries first, here at the top
import numpy as np # math
import matplotlib.pyplot as plt # graphing

#Knowns - what variables did the problem give us? don't forget units!

#in addition to the known values, are there any assumed values?

m = 1 # kg
g = np.array([0, -9.8]) # m / s^2, gravity

pos = np.array([0,0]) # m
v = np.array([20, 38]) # m / s

t = 0 # s
dt = 0.001 # time step

ylist = [pos[1]]
xlist = [pos[0]]
vylist = [v[1]]
vxlist = [v[0]]
tlist = [t]

rho = 1.225 # fluid density kg / m
drag_coeff = 0.47 # drag coefficient
radius = 0.1 # meter
a = radius ** 2 * np.pi # cross sectional area
f_drag_wo_v =  0.5 * rho * drag_coeff * a

#The Motion - Using a while loop, we update the various variables of the object

while ylist[-1] >= 0: #example - while t < 10:
    vmag = np.linalg.norm(v) # calculate magnitude of velocity
    vhat = v / vmag # unit vector

    f_drag = f_drag_wo_v * (vmag ** 2) * -vhat # calculate drag vector 
    
    Fnet = m * g + f_drag
    a = Fnet / m

    v = v + a * dt
    pos = pos + v * dt
    t += dt

    ylist.append(pos[1])
    xlist.append(pos[0])

    vylist.append(v[1])
    vxlist.append(v[0])

    tlist.append(t)


#Graphs - what are the common physics graphs you'd be asked for a question like this? distance vs. time, velocity vs. time

plt.scatter(tlist, ylist, color='blue');   # this makes a scatterplot!
plt.xlabel("Time (s)") # x is time
plt.ylabel("Height (m)") # y is height
plt.title("Ball Height vs. Time") # 

plt.show()

plt.scatter(tlist, xlist, color='blue');   # this makes a scatterplot!
plt.xlabel("Time (s)") # x is time
plt.ylabel("Distance (m)") # y is height
plt.title("Ball Distance vs. Time") # title

plt.show()

plt.scatter(xlist, ylist, color='purple');   # this makes a scatterplot!
plt.xlabel("Distance (s)") # x is time
plt.ylabel("Height (m/s)") # y is velocity
plt.title("Ball Height vs. Distance") # title

plt.show()

plt.scatter(tlist, vylist, color='blue');   # this makes a scatterplot!
plt.xlabel("Time (s)") # x is time
plt.ylabel("Velocity (m)") # y is height
plt.title("Ball Y Velocity vs. Time") # title

plt.show()

plt.scatter(tlist, vxlist, color='blue');   # this makes a scatterplot!
plt.xlabel("Time (s)") # x is time
plt.ylabel("Velocity (m)") # y is height
plt.title("Ball X Velocity vs. Time") # title

plt.show()

np.set_printoptions(precision=3)
#How do we get the computer to tell us the answer? Where is this variable being stored? How do we print it out?
print(f"The ball takes {t} seconds to hit the ground, with a final velocity of {v} meters per second, and travels {xlist[-1]} meters") 
    # show time taken and input height 
```
