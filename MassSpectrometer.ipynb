```python
import matplotlib.pyplot as plt
import numpy as np

# define key constants
m_p = 1.67E-27     # mass of proton: kg
qe =  1.6022E-19      # charge of proton: C

# now, setting up an alpha particle
m = 3.9E-25  #kg
q = qe #C

dt = 2E-9 #s
t = 0 #s starting time

pos = np.array([0, 0, 0]) #m, sets initial position
v = np.array([0, 0, 0]) #m/s, sets initial velocity

#create lists for graphing later
xlist = [pos[0]]
ylist = [pos[1]]
tlist = [t]


# Strength of Fields
B = np.array([0, 0, 0]) #Unit: Teslas, magnetic field is in the z-direction 
E = np.array([5E4, 0, 0]) #unit: N/C, electric field is zero for this example

cross1 = lambda x,y:np.cross(x,y)

# Euler time steps
while pos[0] < 0.05:
    #print(pos)
    Fnet = q * (E + cross1(v, B)) #Defining net force vector
    a = Fnet / m #calculating accerlation vector, Newton's 2nd Law
    v = v + a * dt #updating velocity vector 
    pos = pos + v * dt #updating position vector
    
    t += dt #updating time
    
    #updating lists
    xlist.append(pos[0])
    ylist.append(pos[1])
    tlist.append(t)
    
print(f"Velocity for part A: {np.linalg.norm(v)} m/s")

E = np.array([0, -2E4, 0])
B = np.array([0, 0, 1]) * ( np.linalg.norm(E) / np.linalg.norm(v)) #Unit: Teslas, magnetic field is in the z-direction 
E = np.array([0, 0, 0]) #unit: N/C, electric field is zero for this example

while pos[0] >= 0.05:
    Fnet = q * (E + cross1(v, B)) #Defining net force vector
    a = Fnet / m #calculating accerlation vector, Newton's 2nd Law
    v = v + a * dt #updating velocity vector 
    pos = pos + v * dt #updating position vector
    
    t += dt #updating time

    #updating lists
    xlist.append(pos[0])
    ylist.append(pos[1])
    tlist.append(t)

# Plot the particle's trajectory, in the xy-plane
plt.plot(xlist, ylist)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Particle Trajectoriy in Uniform Magnetic Field")

plt.show()
```

$$ |\vec{v}||\vec{B}|sin\theta = -\vec{E} $$
$$ |\vec{v}||\vec{B}| = |-\vec{E}| $$
$$ |\vec{B}| = \frac{|-\vec{E}| }{ |\vec{v}| } $$
